```python
import matplotlib.pyplot as plt
import numpy as np
name=["arul","Praveen","Hari"]
xaxis=np.arange(len(name))#string cannot be plotted so used as number
Gpa3=[6.9,7.9,8.0]
Gpa4=[7.2,8.0,7.66]
plt.xlabel("NAME")
plt.ylabel("GPA")
plt.title("CGPA ANALYSIS")
plt.xticks(xaxis,name)#to convert number to name in chart
plt.bar(xaxis,Gpa3,label="SEM 3")
plt.bar(xaxis,Gpa4,label="SEM 4")
plt.legend()
```

```python
plt.bar(xaxis+0.1,Gpa3,label="SEM 3",width=0.3)
plt.bar(xaxis-0.2,Gpa4,label="SEM 4",width=0.3)
plt.legend()
```

```python
plt.barh(xaxis+0.1,Gpa3,label="SEM 3")
plt.barh(xaxis-0.2,Gpa4,label="SEM 4")
plt.legend()
```

HISTOGRAM

```python
import matplotlib.pyplot as plt
men_mark=[20,96,45,78,25,65,45,85,12,55,76]
plt.hist(men_mark)
```

```python
import matplotlib.pyplot as plt
men_mark=[20,96,45,78,25,65,45,85,12,55,76]
plt.hist(men_mark,bins=[0,25,50,75,100],rwidth=0.95)
```

```python
women_mark=[45,78,16,69,25,35,78,54,99,94,57]
plt.hist([men_mark,women_mark],bins=[0,25,50,75,100],color=['green','red'],label=["mens","women"])
plt.legend()
plt.title("men-women mark analysis")
```

PIE-CHART

```python
import matplotlib.pyplot as plt
amount=[2000,4000,1000,1500,3000]
need=["phonr","rent","saving","petrol","other"]
plt.pie(amount,labels=need)
plt.show()
```

```python
plt.pie(amount,labels=need,autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(amount,labels=need,autopct='%0.1f%%',explode=[0.1,0,0.2,0,0])
plt.show()
plt.savefig("piechart.png",bbob_inches="tight",pad_inches=2)
```
